### WINSORIZATION

```python
import numpy as np
import pandas as pd

df_winsorized_train = df_application_train_imputed.copy()  # Definisikan df_winsorized_train

# Tentukan kolom numerik pada data train (kecualikan kolom kategorikal yang telah di-encode)
numeric_columns = df_winsorized_train.select_dtypes(include=['float64', 'int64']).columns

# Kolom yang sudah di-encode dan tidak perlu diwinsorizing
exclude_columns = ['TARGET', 'NAME_CONTRACT_TYPE', 'CODE_GENDER', 'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE', 
                   'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 
                   'OCCUPATION_TYPE', 'WEEKDAY_APPR_PROCESS_START', 'ORGANIZATION_TYPE', 
                   'FONDKAPREMONT_MODE', 'HOUSETYPE_MODE', 'WALLSMATERIAL_MODE']

# Menghapus kolom-kolom yang perlu dikecualikan
numeric_columns = [col for col in numeric_columns if col not in exclude_columns]

# Copy data train yang sudah diimputasi
df_winsorized_train = df_application_train_imputed.copy()

# Menghitung IQR untuk setiap kolom numerik
Q1 = df_winsorized_train[numeric_columns].quantile(0.25)  # Kuartil pertama
Q3 = df_winsorized_train[numeric_columns].quantile(0.75)  # Kuartil ketiga
IQR = Q3 - Q1  # Menghitung rentang interkuartil

# Menentukan batas bawah dan atas untuk mendeteksi outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menerapkan Winsorizing pada setiap kolom numerik (kecuali yang dikecualikan)
for col in numeric_columns:
    df_winsorized_train[col] = np.where(df_winsorized_train[col] < lower_bound[col], lower_bound[col], df_winsorized_train[col])
    df_winsorized_train[col] = np.where(df_winsorized_train[col] > upper_bound[col], upper_bound[col], df_winsorized_train[col])

# Menampilkan hasil setelah Winsorizing
print(df_winsorized_train.head())

# Menghitung IQR untuk setiap kolom numerik setelah Winsorizing
Q1_winsorized = df_winsorized_train[numeric_columns].quantile(0.25)  # Kuartil pertama
Q3_winsorized = df_winsorized_train[numeric_columns].quantile(0.75)  # Kuartil ketiga
IQR_winsorized = Q3_winsorized - Q1_winsorized  # Menghitung rentang interkuartil

# Menentukan batas bawah dan atas untuk mendeteksi outlier setelah Winsorizing
lower_bound_winsorized = Q1_winsorized - 1.5 * IQR_winsorized
upper_bound_winsorized = Q3_winsorized + 1.5 * IQR_winsorized

# Menyaring outlier setelah Winsorizing
outliers_winsorized = (df_winsorized_train[numeric_columns] < lower_bound_winsorized) | (df_winsorized_train[numeric_columns] > upper_bound_winsorized)

# Menampilkan jumlah outlier per kolom setelah Winsorizing
outlier_count_winsorized = outliers_winsorized.sum()
print(f"Jumlah outlier per kolom setelah Winsorizing:\n{outlier_count_winsorized}")

# Menampilkan kolom yang memiliki outlier setelah Winsorizing
outlier_columns_winsorized = outlier_count_winsorized[outlier_count_winsorized > 0].index
print(f"Kolom yang memiliki outlier setelah Winsorizing: {list(outlier_columns_winsorized)}")
```

### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, log_loss, roc_auc_score
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import numpy as np

# Buat salinan dari df_winsorized_train untuk digunakan dalam pelatihan Logistic Regression
df_winsorized_train_logreg = df_winsorized_train.copy()  # Salinan untuk Logistic Regression

# Memisahkan fitur dan target dari data train (df_winsorized_train) untuk Logistic Regression
X_train_logreg = df_winsorized_train_logreg.drop(columns=['TARGET', 'SK_ID_CURR'], errors='ignore')  # Fitur
y_train_logreg = df_winsorized_train_logreg['TARGET']  # Target
X_test = df_application_test_imputed.drop(columns=['SK_ID_CURR'], errors='ignore')  # Data test tanpa target

# 1. Feature Scaling: Scaling fitur numerik menggunakan StandardScaler
scaler_logreg = StandardScaler()
X_train_scaled_logreg = scaler_logreg.fit_transform(X_train_logreg)
X_test_scaled = scaler_logreg.transform(X_test)  # Apply scaling to test data using the same scaler

# 2. Data Balancing: Menggunakan SMOTE untuk oversampling
smote_logreg = SMOTE(random_state=42)
X_train_balanced_logreg, y_train_balanced_logreg = smote_logreg.fit_resample(X_train_scaled_logreg, y_train_logreg)

# 3. Membuat model Logistic Regression
logreg_model = LogisticRegression(random_state=42, solver='liblinear', class_weight='balanced')  # 'balanced' untuk menangani ketidakseimbangan kelas

# 4. Stratified K-Fold Cross-Validation
cv_logreg = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 5. Menghitung skor cross-validation secara manual
accuracies_logreg = []
log_losses_logreg = []
roc_aucs_logreg = []
for train_index, val_index in cv_logreg.split(X_train_balanced_logreg, y_train_balanced_logreg):
    # Membagi data menjadi train dan validation set
    X_train_cv_logreg, X_val_cv_logreg = X_train_balanced_logreg[train_index], X_train_balanced_logreg[val_index]
    y_train_cv_logreg, y_val_cv_logreg = y_train_balanced_logreg[train_index], y_train_balanced_logreg[val_index]
    
    # Melatih model pada setiap fold
    logreg_model.fit(X_train_cv_logreg, y_train_cv_logreg)
    
    # Melakukan prediksi pada validation set
    y_val_pred_logreg = logreg_model.predict(X_val_cv_logreg)
    y_val_pred_prob_logreg = logreg_model.predict_proba(X_val_cv_logreg)[:, 1]  # Mengambil probabilitas untuk kelas 1
    
    # Menghitung akurasi pada fold ini
    acc_logreg = accuracy_score(y_val_cv_logreg, y_val_pred_logreg)
    accuracies_logreg.append(acc_logreg)
    
    # Menghitung log loss
    logloss_logreg = log_loss(y_val_cv_logreg, y_val_pred_prob_logreg)
    log_losses_logreg.append(logloss_logreg)
    
    # Menghitung AUC (Area Under Curve)
    auc_logreg = roc_auc_score(y_val_cv_logreg, y_val_pred_prob_logreg)
    roc_aucs_logreg.append(auc_logreg)

# Menampilkan hasil cross-validation untuk Logistic Regression
print(f"Cross-validation accuracy scores (Logistic Regression): {accuracies_logreg}")
print(f"Mean accuracy (Logistic Regression): {np.mean(accuracies_logreg):.4f}")
print(f"Mean log loss (Logistic Regression): {np.mean(log_losses_logreg):.4f}")
print(f"Mean AUC (Logistic Regression): {np.mean(roc_aucs_logreg):.4f}")
```

```python
# 6. Melatih model pada seluruh data train (setelah cross-validation) dan prediksi pada data test
logreg_model.fit(X_train_scaled_logreg, y_train_logreg)  # Melatih model pada seluruh data train

# 7. Melakukan prediksi pada data test
y_test_pred_logreg = logreg_model.predict(X_test_scaled)

# 8. Menyimpan hasil prediksi dalam kolom TARGET pada df_application_test_imputed
df_application_test_imputed['TARGET'] = y_test_pred_logreg

# Menyimpan hasil prediksi ke dalam file CSV di /kaggle/working
file_name_logreg = '/kaggle/working/submission_logreg_predictions.csv'  # Path untuk menyimpan di Kaggle environment
df_application_test_imputed[['SK_ID_CURR', 'TARGET']].to_csv(file_name_logreg, index=False)

# Menampilkan beberapa baris dari hasil prediksi
print(f"Hasil prediksi Logistic Regression telah disimpan ke dalam file {file_name_logreg}")
print(df_application_test_imputed[['SK_ID_CURR', 'TARGET']].head())
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, log_loss, roc_auc_score
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import numpy as np

# Buat salinan dari df_winsorized_train untuk digunakan dalam pelatihan XGBoost
df_winsorized_train_xgb = df_winsorized_train.copy()  # Salinan untuk XGBoost

# Memisahkan fitur dan target dari data train (df_winsorized_train) untuk XGBoost
X_train_XGB = df_winsorized_train_xgb.drop(columns=['TARGET', 'SK_ID_CURR'], errors='ignore')  # Fitur
y_train_XGB = df_winsorized_train_xgb['TARGET']  # Target
X_test_XGB = df_application_test_imputed.drop(columns=['SK_ID_CURR'], errors='ignore')  # Data test tanpa target

# Pastikan kolom pada X_test sama dengan X_train, jika ada kolom yang hilang, tambahkan kolom kosong
X_test_XGB = X_test_XGB[X_train_XGB.columns]

# 1. Feature Scaling: Scaling fitur numerik menggunakan StandardScaler
scaler_XGB = StandardScaler()
X_train_scaled_XGB = scaler_XGB.fit_transform(X_train_XGB)
X_test_scaled_XGB = scaler_XGB.transform(X_test_XGB)  # Apply scaling to test data using the same scaler

# 2. Data Balancing: Menggunakan SMOTE untuk oversampling
smote_XGB = SMOTE(random_state=42)
X_train_balanced_XGB, y_train_balanced_XGB = smote_XGB.fit_resample(X_train_scaled_XGB, y_train_XGB)

# 3. Membuat model XGBoost
xgb_model = XGBClassifier(random_state=42, scale_pos_weight=2)  # Menambahkan parameter scale_pos_weight untuk menangani ketidakseimbangan kelas

# 4. Stratified K-Fold Cross-Validation
cv_XGB = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 5. Menghitung skor cross-validation secara manual
accuracies_XGB = []
log_losses_XGB = []
roc_aucs_XGB = []
for train_index, val_index in cv_XGB.split(X_train_balanced_XGB, y_train_balanced_XGB):
    # Membagi data menjadi train dan validation set
    X_train_cv_XGB, X_val_cv_XGB = X_train_balanced_XGB[train_index], X_train_balanced_XGB[val_index]
    y_train_cv_XGB, y_val_cv_XGB = y_train_balanced_XGB[train_index], y_train_balanced_XGB[val_index]
    
    # Melatih model pada setiap fold
    xgb_model.fit(X_train_cv_XGB, y_train_cv_XGB)
    
    # Melakukan prediksi pada validation set
    y_val_pred_XGB = xgb_model.predict(X_val_cv_XGB)
    y_val_pred_prob_XGB = xgb_model.predict_proba(X_val_cv_XGB)[:, 1]  # Mengambil probabilitas untuk kelas 1
    
    # Menghitung akurasi pada fold ini
    acc_XGB = accuracy_score(y_val_cv_XGB, y_val_pred_XGB)
    accuracies_XGB.append(acc_XGB)
    
    # Menghitung log loss
    logloss_XGB = log_loss(y_val_cv_XGB, y_val_pred_prob_XGB)
    log_losses_XGB.append(logloss_XGB)
    
    # Menghitung AUC (Area Under Curve)
    auc_XGB = roc_auc_score(y_val_cv_XGB, y_val_pred_prob_XGB)
    roc_aucs_XGB.append(auc_XGB)

# Menampilkan hasil cross-validation untuk XGBoost
print(f"Cross-validation accuracy scores (XGB): {accuracies_XGB}")
print(f"Mean accuracy (XGB): {np.mean(accuracies_XGB):.4f}")
print(f"Mean log loss (XGB): {np.mean(log_losses_XGB):.4f}")
print(f"Mean AUC (XGB): {np.mean(roc_aucs_XGB):.4f}")
```

```python
# Melatih model pada seluruh data train dan prediksi pada data test
xgb_model.fit(X_train_scaled_XGB, y_train_XGB)
y_test_pred_XGB = xgb_model.predict(X_test_scaled_XGB)

# Menyimpan hasil prediksi dalam kolom TARGET pada df_application_test_imputed
df_application_test_imputed['TARGET'] = y_test_pred_XGB

# Menyimpan hasil prediksi ke dalam file CSV di /kaggle/working
file_name_XGB = '/kaggle/working/submission_xgb_predictions.csv'  # Path untuk menyimpan di Kaggle environment
df_application_test_imputed[['SK_ID_CURR', 'TARGET']].to_csv(file_name_XGB, index=False)

# Menampilkan beberapa baris dari hasil prediksi
print(f"Hasil prediksi XGBoost telah disimpan ke dalam file {file_name_XGB}")
print(df_application_test_imputed[['SK_ID_CURR', 'TARGET']].head())
```
